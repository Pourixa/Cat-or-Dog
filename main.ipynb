```python
import pandas as pd
import numpy as np
import tensorflow as tf
import skimage as ski
import matplotlib.pyplot as plt
import os
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense ,Conv2D,MaxPooling2D,Flatten,BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("bhavikjikadara/dog-and-cat-classification-dataset")

print("Path to dataset files:", path)
```

```python
def toGrayScaleArray(imgPath):
    img=ski.transform.resize(ski.io.imread(imgPath,as_gray=True),(120,120))
    if img.shape == (120,120):
        return img
```

```python
fileCat = os.listdir(r"/kaggle/input/dog-and-cat-classification-dataset/PetImages/Cat")
fileDog = os.listdir(r"/kaggle/input/dog-and-cat-classification-dataset/PetImages/Dog")
```

```python
dataCat = np.zeros((len(fileCat),14401))
for i in range(len(fileCat)) :
    try:
        cat=toGrayScaleArray(fr"/kaggle/input/dog-and-cat-classification-dataset/PetImages/Cat/{fileCat[i]}")
        cat = np.append(cat,0)
        dataCat[i] = cat
    except Exception as e:
        print(e)
        continue
del fileCat
```

```python
dataCat=dataCat[~np.all(dataCat == 0, axis=1)]
```

```python
dataDog = np.zeros((len(fileDog),14401))
for i in range(len(fileDog)):
    try:
        dog=toGrayScaleArray(rf"/kaggle/input/dog-and-cat-classification-dataset/PetImages/Dog/{fileDog[i]}")
        dog = np.append(dog,1)
        dataDog[i] = dog
    except Exception as e:
        print(e)
        continue
del fileDog
```

```python
dataDog=dataDog[~np.all(dataDog == 0, axis=1)]
```

```python
sixty = int(len(dataCat)*0.6)
trainCat = dataCat[0:sixty]
_cat = dataCat[sixty:-1]
del dataCat
catCv = _cat[0:int(len(_cat)*0.5)]
catTest = _cat[int(len(_cat)*0.5):-1]
del _cat

```

```python
sixty = int(len(dataDog)*0.6)
trainDog = dataDog[0:sixty]
_dog = dataDog[sixty:-1]
del dataDog
dogCv = _dog[0:int(len(_dog)*0.5)]
dogTest = _dog[int(len(_dog)*0.5):-1]
del _dog
```

```python
model = Sequential([
    tf.keras.Input((120,120,1)),
    Conv2D(16,(4,4),activation="relu",),
    BatchNormalization(),
    MaxPooling2D((2,2),strides = 2,),
    Conv2D(32,(2,2),activation="relu",),
    BatchNormalization(),
    MaxPooling2D((2,2),strides = 2,),
    Conv2D(64,(2,2),activation="relu"),
    BatchNormalization(),
    MaxPooling2D((2,2),strides = 2),
    Conv2D(128,(2,2),activation="relu"),
    BatchNormalization(),
    MaxPooling2D((2,2),strides = 2),
    Flatten(),
    Dense(18,activation="relu"),
    Dense(8,activation="relu"),
    Dense(1,activation="sigmoid")
    ]

)
```

```python
model.compile(loss=tf.losses.BinaryCrossentropy(),optimizer=tf.optimizers.Adam(learning_rate=0.001),metrics=["accuracy",])
```

```python
model.summary()
```

```python
dataTrain = np.vstack((trainCat,trainDog))
Xtrain = dataTrain[:,:14400]
Ytrain = dataTrain[:,-1]
```

```python
del dataTrain
```

```python
dataCv = np.vstack((catCv,dogCv))
XCv = dataCv[:,:14400]
YCv = dataCv[:,-1]
```

```python
del dataCv
```

```python
dataTest = np.vstack((catTest,dogTest))
Xtest = dataTest[:,:14400]
Ytest = dataTest[:,-1]
```

```python
del dataTest
```

```python
del catCv,catTest,trainCat,dogCv,dogTest,trainDog
```

```python
del model
```

```python
ES = EarlyStopping(patience=3,restore_best_weights=True)
```

```python
model.fit(Xtrain.reshape((-1,120,120,1)),Ytrain,epochs=50,validation_data=(XCv.reshape((-1,120,120,1)),YCv),callbacks=(ES))
```

```python
model.save("MODEL3.keras")
```
